```python
#dependent and independent features

X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
type(X)
```

```python
#splitting the data

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
type(X_train)
```

```python
#standardization

from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
#model fitting 

from sklearn.linear_model import LinearRegression

model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
#metrics

from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

print("mean absolute error is",mean_absolute_error(y_test,y_pred))
print("mean squared error is",mean_squared_error(y_test,y_pred))
print("root mean square is",np.sqrt(mean_squared_error(y_test,y_pred)))
r2_score=r2_score(y_test,y_pred)
print("r2_score",r2_score)
```

```python
#adj r2

n=len(y_test)
p=X_test.shape[1]
adjusted_r2=1-(1-r2_score)*(n-1)/(n-p-1)
print("the adjusted r^2 is",adjusted_r2)
```

```python
## Assumptions:

plt.scatter(y_test,y_pred)
plt.xlabel('y_test')
plt.ylabel('y_pred')
```

```python
#residuals

residuals=y_test-y_pred

residuals
```

```python
sns.displot(residuals,kind='kde')
```

```python
#multicollinearity

plt.scatter(y_pred,residuals)  #if its uniformally distributed there is no multicollinearity

```
